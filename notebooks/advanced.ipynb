## 自然言語処理

```python
json_path = Path("DBDC2_dev/DCM/1470623130.log.json")
json_path.exists()
```

```python
import json
dialogue = json.load(json_path.open(mode="r"))
dialogue
```

```python
utterance = dialogue["turns"][0]["utterance"]
utterance
```

```python
from ginza import *
import spacy
nlp = spacy.load("ja_ginza")
```

```python
doc = nlp(utterance)
spacy.displacy.render(doc, style='dep', jupyter=True)
```

## 可視化

```python
import seaborn as sns
```

```python
utterance_lens = []
for json_path in data_dir.glob("*/*.log.json"):
    dialogue = json.load(json_path.open(mode="r"))
    for turn in dialogue["turns"]:
        utterance_lens.append(len(turn["utterance"]))
        
sns.distplot(utterance_lens)
```
